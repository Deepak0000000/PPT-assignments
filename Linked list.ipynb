```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def remove_elements(head, val):
   
    while head and head.value == val:
        head = head.next

    current = head
    while current and current.next:
        
        if current.next.value == val:
            current.next = current.next.next
        else:
            current = current.next

    return head


def create_linked_list(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head


def print_linked_list(head):
    current = head
    while current:
        print(current.value, end=" -> ")
        current = current.next
    print("None")


if __name__ == "__main__":
    
    head = create_linked_list([1, 2, 6, 3, 4, 5, 6])

    print("Original linked list:")
    print_linked_list(head)

    val_to_remove = 6
    new_head = remove_elements(head, val_to_remove)

    print("\nLinked list after removing elements with value {}:".format(val_to_remove))
    print_linked_list(new_head)
```

```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def merge_two_lists(list1, list2):
   
    dummy = ListNode(-1)
    current = dummy


    while list1 and list2:
        if list1.value <= list2.value:
            current.next = list1
            list1 = list1.next
        else:
            current.next = list2
            list2 = list2.next
        current = current.next

    
    current.next = list1 if list1 else list2

    return dummy.next


def create_linked_list(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head


def print_linked_list(head):
    current = head
    while current:
        print(current.value, end=" -> ")
        current = current.next
    print("None")


if __name__ == "__main__":
 
    list1 = create_linked_list([1, 2, 4])
    list2 = create_linked_list([1, 3, 4])

    print("List 1:")
    print_linked_list(list1)

    print("\nList 2:")
    print_linked_list(list2)

    merged_list = merge_two_lists(list1, list2)

    print("\nMerged Linked List:")
    print_linked_list(merged_list)
```

```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def detect_cycle(head):
    if not head or not head.next:
        return None

    # Function to check if there is a cycle and find the meeting point
    def find_meeting_point():
        slow = head
        fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return slow
        return None

    # Find the meeting point where slow and fast pointers meet
    meeting_point = find_meeting_point()

    # If there is no cycle, return None
    if not meeting_point:
        return None

    # Reset one of the pointers to the head
    slow = head

    # Move both pointers one step at a time until they meet again
    while slow != meeting_point:
        slow = slow.next
        meeting_point = meeting_point.next

    # The node where they meet again is the starting point of the cycle
    return meeting_point


def create_linked_list_with_cycle(arr, pos):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    cycle_node = None
    for i in range(1, len(arr)):
        current.next = ListNode(arr[i])
        current = current.next
        if i == pos:
            cycle_node = current
    if cycle_node:
        current.next = cycle_node
    return head


def print_linked_list(head):
    current = head
    while current:
        print(current.value, end=" -> ")
        current = current.next
    print("None")


if __name__ == "__main__":
   
    head = create_linked_list_with_cycle([3, 2, 0, -4], 1)

    print("Linked List with cycle:")
    print_linked_list(head)

    cycle_start_node = detect_cycle(head)

    if cycle_start_node:
        print("\nCycle starts at node with value:", cycle_start_node.value)
    else:
        print("\nNo cycle in the linked list.")
```

```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def left_rotate_linked_list(head, k):
    if not head or k <= 0:
        return head

    # Find the length of the linked list
    length = 1
    current = head
    while current.next:
        length += 1
        current = current.next

    # Calculate the effective rotation amount (k % length)
    k = k % length

    if k == 0:
        return head  # No rotation needed

    # Create a dummy node as the starting point of the linked list
    dummy = ListNode(-1)
    dummy.next = head

    first = dummy
    second = dummy

    # Move the first pointer k steps ahead
    for _ in range(k):
        first = first.next

    # Move both pointers until the first pointer reaches the end
    while first.next:
        first = first.next
        second = second.next

    # Update the pointers to achieve the rotation
    dummy.next = second.next
    second.next = None
    first.next = head

    return dummy.next

# Function to create a linked list from a list
def create_linked_list(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.value, end=" -> ")
        current = current.next
    print("None")

# Test the left_rotate_linked_list function
if __name__ == "__main__":
    # Create the linked list: 2 -> 4 -> 7 -> 8 -> 9
    head = create_linked_list([2, 4, 7, 8, 9])
    k = 3

    print("Original linked list:")
    print_linked_list(head)

    new_head = left_rotate_linked_list(head, k)

    print("\nLinked list after left-rotating by {} nodes:".format(k))
    print_linked_list(new_head)
```

```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def delete_consecutive_sum_zero(head):
    dummy = ListNode(0)
    dummy.next = head

    prefix_sum = 0
    sum_to_node = {0: dummy}


    current = dummy.next
    while current:
        prefix_sum += current.value
        sum_to_node[prefix_sum] = current
        current = current.next

   
    current = dummy.next
    prefix_sum = 0
    while current:
        prefix_sum += current.value
        current.next = sum_to_node[prefix_sum].next
        current = current.next

    return dummy.next


def create_linked_list(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head


def print_linked_list(head):
    current = head
    while current:
        print(current.value, end=" -> ")
        current = current.next
    print("None")


if __name__ == "__main__":
  
    head = create_linked_list([1, 2, -3, 3, 1])

    print("Original linked list:")
    print_linked_list(head)

    new_head = delete_consecutive_sum_zero(head)

    print("\nLinked list after deleting consecutive sequences with sum 0:")
    print_linked_list(new_head)
```

```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def odd_even_linked_list(head):
    if not head or not head.next:
        return head

    odd_head = head
    even_head = head.next

    odd_current = odd_head
    even_current = even_head

    while even_current and even_current.next:
        odd_current.next = even_current.next
        odd_current = odd_current.next
        even_current.next = odd_current.next
        even_current = even_current.next

    odd_current.next = even_head

    return odd_head

def create_linked_list(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head


def print_linked_list(head):
    current = head
    while current:
        print(current.value, end=" -> ")
        current = current.next
    print("None")


if __name__ == "__main__":

    head = create_linked_list([1, 2, 3, 4, 5])

    print("Original linked list:")
    print_linked_list(head)

    new_head = odd_even_linked_list(head)

    print("\nLinked list after reordering:")
    print_linked_list(new_head)
```
